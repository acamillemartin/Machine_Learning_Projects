## **K Means Clustering Creation**

We now proceed on to find the optimum number of clusters in the given dataset.
We will only use the 4 columns from the dataset as it unsupervised learning

```python
# selecting the feaures: sepal length (cm)	| sepal width (cm) |petal length (cm)	| petal width (cm)
X = iris_df.iloc[:, :4].values
```

### Specify first the number of clusters or the k

To find optimum number of clusters, we use the Elbow Method. The Elbow method runs K-means clustering on a dataset for a range of values and then for each value of k, computes an average score. The Elbow Method is the most popular method to determine the optimal number of clusters into which the data may be clustered! For each value, within-cluster sum of squares value is computed and stored in some list. These values are then plotted against the number of clusters used before. The location of bend in the graph indicates the optimal number of clusters.

From the above graph we choose the number of clusters as **3** as the optimum clusters. We can now apply k-mean clustering on the data

```python
# We are going to use the imported Kmeans library above

# Applying kmeans to the dataset / Creating the kmeans classifier
kmeans = KMeans(n_clusters=3, init='k-means++',
                max_iter=300, n_init=10, random_state=0)

# max_iter represent the maximum number ot iteration for the k-means algorithm
# n_init is the no of times the k-means algo will be run with difference centrod seeds

 # fit_predict() is mainly used in unsupervised machine learning where we dont have labelled inputs
y_pred = kmeans.fit_predict(X)
```

### Visualization of the Clusters

```python
# Visualising the clusters
plt.scatter(X[y_pred == 0, 0], X[y_pred == 0, 1], s = 100, c = 'red', label = 'iris_setosa')
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], s = 100, c = 'blue', label = 'iris_versicolour')
plt.scatter(X[y_pred == 2, 0], X[y_pred == 2, 1], s = 100, c = 'green', label = 'iris_virginica')

# Plotting the centroids of the clusters which represents the imaginary or real location of center of the cluster
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')
plt.title("K-Means Cluster")
plt.legend()
plt.show()
```

With the above, we have predicted the optimal number of clusters and represented it visually using K-Mean Clustering Algorithm
